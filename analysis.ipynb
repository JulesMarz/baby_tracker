```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
df = pd.read_excel('data/tracking.xlsx')
# To date time 
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['hour'] = df['Timestamp'].dt.hour
df['day'] = df['Timestamp'].dt.day
df['month'] = df['Timestamp'].dt.month
df['date'] = pd.to_datetime(df['Timestamp'].dt.date)
```

```python
# Plot activities
aux = df.groupby('Activity')['Timestamp'].count()
plt.figure(figsize=(10,5))
plt.title('Total Count of Activities')
plt.bar(aux.index, aux.values)
plt.savefig('figures/total_count.png')
plt.show()
```

```python
# Number of times 'Go to sleep' each day
plt.figure(figsize=(10,5))
plt.title('Count of Go to sleep')
# If you change the 'Go to sleep' activity plots every activity
df[df['Activity']=='Go to sleep'].groupby('date')['Activity'].count().plot()
plt.ylabel('Count')
plt.xlabel('')
plt.xticks(rotation=45)
plt.savefig('figures/bottle_feed.png')
plt.show()

```
